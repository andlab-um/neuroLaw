```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import numpy as np
import os
from pathlib import Path

# Set style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['font.size'] = 10
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['xtick.labelsize'] = 11
plt.rcParams['ytick.labelsize'] = 11

# Set file path
data_dir = Path('./jiamin/fMRI_data/ROI/ROI/8')

# Check if directory exists
if not data_dir.exists():
    print(f"Error: Directory does not exist - {data_dir.absolute()}")
    print("Please check if the path is correct")
    exit()

# Get specific xlsx files (8-1.xlsx to 8-6.xlsx)
xlsx_files = []
for i in range(1, 7):
    file_path = data_dir / f'8-{i}.xlsx'
    if file_path.exists():
        xlsx_files.append(file_path)
    else:
        print(f"Warning: File not found - {file_path.name}")

if len(xlsx_files) == 0:
    print(f"Warning: No Excel files found in {data_dir.absolute()}")
    print("All files in directory:")
    for f in data_dir.iterdir():
        print(f"  - {f.name}")
    exit()

print(f"Found {len(xlsx_files)} Excel files:")
for f in xlsx_files:
    print(f"  - {f.name}")

# Create output directory
output_dir = Path('./output_figures')
output_dir.mkdir(exist_ok=True)

# Process each file
for file_path in xlsx_files:
    print(f"\n{'='*60}")
    print(f"Processing file: {file_path.name}")
    print(f"{'='*60}")
    
    try:
        # Read data
        df = pd.read_excel(file_path, engine='openpyxl')
            
    except Exception as e:
        print(f"Error reading file: {e}")
        print(f"Trying with default engine...")
        try:
            df = pd.read_excel(file_path)
        except Exception as e2:
            print(f"Still unable to read file: {e2}")
            continue
    
    # Get column names
    columns = df.columns.tolist()
    print(f"Columns: {columns}")
    
    # Check data
    if len(columns) < 3:
        print(f"Warning: Less than 3 columns, skipping this file")
        continue
    
    # Assume first column is Subject, second and third are current and delay conditions
    subject_col = columns[0]
    current_col = columns[1]
    delay_col = columns[2]
    
    # Remove missing values
    df = df.dropna(subset=[subject_col, current_col, delay_col])
    
    if len(df) == 0:
        print(f"Warning: No data after removing missing values, skipping this file")
        continue
    
    # Display basic data information
    print(f"\nData preview:")
    print(df.head())
    print(f"\nData shape: {df.shape}")
    print(f"\nNumber of subjects: {len(df)}")
    
    # Get data for paired comparison
    current_data = df[current_col].values
    delay_data = df[delay_col].values
    
    # Extract condition names (remove prefix before underscore if exists)
    current_name = current_col.split('_')[-1] if '_' in current_col else current_col
    delay_name = delay_col.split('_')[-1] if '_' in delay_col else delay_col
    
    # Calculate descriptive statistics
    print(f"\nDescriptive statistics:")
    print(f"\n{current_name} condition:")
    print(f"  Sample size: {len(current_data)}")
    print(f"  Mean: {np.mean(current_data):.4f}")
    print(f"  Std: {np.std(current_data, ddof=1):.4f}")
    print(f"  Median: {np.median(current_data):.4f}")
    print(f"  Min: {np.min(current_data):.4f}")
    print(f"  Max: {np.max(current_data):.4f}")
    
    print(f"\n{delay_name} condition:")
    print(f"  Sample size: {len(delay_data)}")
    print(f"  Mean: {np.mean(delay_data):.4f}")
    print(f"  Std: {np.std(delay_data, ddof=1):.4f}")
    print(f"  Median: {np.median(delay_data):.4f}")
    print(f"  Min: {np.min(delay_data):.4f}")
    print(f"  Max: {np.max(delay_data):.4f}")
    
    # Paired samples t-test
    t_stat, t_p = stats.ttest_rel(current_data, delay_data)
    print(f"\nPaired samples t-test:")
    print(f"  t={t_stat:.4f}, p={t_p:.4f}")
    
    # Effect size (Cohen's d for paired samples)
    diff = current_data - delay_data
    cohens_d = np.mean(diff) / np.std(diff, ddof=1)
    print(f"\nCohen's d (effect size): {cohens_d:.4f}")
    
    # Determine significance marker
    if t_p < 0.001:
        sig_marker = '***'
        sig_text = 'p < 0.001'
    elif t_p < 0.01:
        sig_marker = '**'
        sig_text = f'p = {t_p:.3f}'
    elif t_p < 0.05:
        sig_marker = '*'
        sig_text = f'p = {t_p:.3f}'
    else:
        sig_marker = 'n.s.'
        sig_text = f'p = {t_p:.3f}'
    
    # Create figure with clean styling
    fig, ax = plt.subplots(figsize=(4.5, 6))
    fig.patch.set_facecolor('white')
    ax.set_facecolor('white')
    
    # Set colors - modern palette
    colors = ['#3498db', '#e74c3c']
    
    # Prepare data for plotting
    groups = [current_name, delay_name]
    plot_data = [current_data, delay_data]
    positions = np.arange(len(groups))
    
    # Create violin plot
    parts = ax.violinplot(plot_data, positions=positions, widths=0.7,
                          showmeans=False, showmedians=False, showextrema=False)
    
    # Color the violin plots
    for i, pc in enumerate(parts['bodies']):
        pc.set_facecolor(colors[i])
        pc.set_alpha(0.4)
        pc.set_edgecolor(colors[i])
        pc.set_linewidth(2)
    
    # Add scatter points with jitter
    np.random.seed(42)
    for i, data in enumerate(plot_data):
        x = np.random.normal(i, 0.04, size=len(data))
        ax.scatter(x, data, alpha=0.6, s=50, color=colors[i], 
                  edgecolors='white', linewidth=1.5, zorder=3)
    
    # # Add lines connecting paired data points
    # for j in range(len(current_data)):
    #     ax.plot([0, 1], [current_data[j], delay_data[j]], 
    #            color='gray', alpha=0.3, linewidth=0.8, zorder=1)
    
    # Add mean markers in gray
    means = [np.mean(current_data), np.mean(delay_data)]
    ax.scatter(positions, means, color='#5d6d7e', s=200, marker='D', 
              edgecolors='white', linewidth=2, zorder=4, label='Mean')
    
    # Add median line
    medians = [np.median(current_data), np.median(delay_data)]
    for i, median in enumerate(medians):
        ax.hlines(median, i - 0.35, i + 0.35, colors='#2c3e50', 
                 linewidth=3, zorder=4, alpha=0.8)
    
    # Add error bars (standard error) in gray
    for i, data in enumerate(plot_data):
        mean = np.mean(data)
        sem = stats.sem(data)
        ax.errorbar(i, mean, yerr=sem, fmt='none', ecolor='#5d6d7e', 
                   capsize=10, capthick=2.5, linewidth=2.5, alpha=0.8, zorder=4)
    
    # # Add significance marker
    # y_max = max(np.max(current_data), np.max(delay_data))
    # y_min = min(np.min(current_data), np.min(delay_data))
    # y_range = y_max - y_min
    
    # # Draw significance bracket
    # bracket_y = y_max + y_range * 0.05
    # ax.plot([0, 0, 1, 1], [bracket_y, bracket_y + y_range * 0.02, 
    #         bracket_y + y_range * 0.02, bracket_y], 
    #        color='#2c3e50', linewidth=2, zorder=5)
    
    # # Add significance text
    # ax.text(0.5, bracket_y + y_range * 0.04, sig_marker, 
    #        ha='center', va='bottom', fontsize=18, fontweight='bold', color='#2c3e50')
    # ax.text(0.5, bracket_y + y_range * 0.08, sig_text, 
    #        ha='center', va='bottom', fontsize=11, color='#2c3e50')
    
    # Set labels and title
    ax.set_xticks(positions)
    ax.set_xticklabels(groups, fontsize=13, fontweight='bold')
    
    # Extract ROI name from column names (part before underscore)
    roi_name = current_col.split('_')[0] if '_' in current_col else 'ROI'
    
    ax.set_ylabel(f'{roi_name} Activation', fontsize=14, fontweight='bold', color='#2c3e50')
    ax.set_xlabel('Condition', fontsize=14, fontweight='bold', color='#2c3e50')
    
    # Set title
    file_name_without_ext = file_path.stem
    ax.set_title(f'{file_name_without_ext}', 
                fontsize=16, fontweight='bold', pad=20, color='#2c3e50')
    
    # Add grid
    ax.grid(True, alpha=0.3, linestyle='--', axis='y', color='#bdc3c7', linewidth=1)
    ax.set_axisbelow(True)
    
    # Adjust y-axis limits to accommodate significance marker
    # ax.set_ylim(y_min - y_range * 0.05, bracket_y + y_range * 0.15)
    
    # Remove top and right spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_linewidth(2)
    ax.spines['bottom'].set_linewidth(2)
    ax.spines['left'].set_color('#2c3e50')
    ax.spines['bottom'].set_color('#2c3e50')
    
    # Adjust layout
    plt.tight_layout()
    
    # Save figure
    output_path = output_dir / f'{file_name_without_ext}_paired.png'
    plt.savefig(output_path, dpi=300, bbox_inches='tight', facecolor='white', edgecolor='none')
    print(f"\nFigure saved to: {output_path}")
    
    plt.show()
    plt.close()

print(f"\n{'='*60}")
print("All files processed!")
print(f"Figures saved in: {output_dir.absolute()}")
print(f"{'='*60}")
```
